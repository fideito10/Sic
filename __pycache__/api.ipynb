```python
# =============================================================================
# CELDA 1: FUNCIONES AUXILIARES COMPLETAS
# =============================================================================

# =============================================================================
# FUNCIONES BÃSICAS DE API
# =============================================================================

def simple_api_test():
    """MÃ©todo simple que replica tu cÃ³digo que funciona"""
    print("ðŸ§ª Probando conexiÃ³n al API...")
    
    BASE_URL = "https://connect-us.catapultsports.com/api/v6"
    
    # Leer el token desde el archivo
    API_KEY = 'eyJ0eXAiOiJKV1QiLCJhbGciOiJSUzI1NiIsImtpZCI6IjEzMWY5NGIxOTg3ZGY4NzcxNTljOGQ2MTAzMTIzNDNjIn0.eyJhdWQiOiI0NjFiMTExMS02ZjdhLTRkYmItOWQyOS0yMzAzOWZlMjI4OGUiLCJqdGkiOiIyNzFlM2U1MjMzNDRiMjAxOTQ4MWViYmJkNjk2NTJkZDMwNmZiZThiMjU5NDkxZGI5ZGRhYjJiMWQ1Y2ViYmJjOWVlYzFiODFiYzNiNzdlNyIsImlhdCI6MTc1MzExNjQ1NS45MDQ2MjYsIm5iZiI6MTc1MzExNjQ1NS45MDQ2MjgsImV4cCI6NDkwNjcxNjQ1NS44OTQ1ODMsInN1YiI6ImJkODAyMzAxLTk1YzgtNDgxNy05ZTAxLWFjOTI5Y2ZlZGMwNiIsInNjb3BlcyI6WyJjb25uZWN0IiwiY2F0YXB1bHRyIiwic2Vuc29yLXJlYWQtb25seSJdLCJpc3MiOiJodHRwczovL2JhY2tlbmQtdXMub3BlbmZpZWxkLmNhdGFwdWx0c3BvcnRzLmNvbSIsImNvbS5jYXRhcHVsdHNwb3J0cyI6eyJvcGVuZmllbGQiOnsiY3VzdG9tZXJzIjpbeyJyZWxhdGlvbiI6ImF1dGgiLCJpZCI6MTM3N31dfX19.P-w_Y3suqBxKtJcMMlaJEZBI-2z8BLOOkRhwVu85W5EG9y2MCeLukR4zMmr4G7bvEbTchaNCPXgEE4FSbuz7xGEFZVsLTyeYb6drEClcgY3SvXPNCH4Y3iwk93ioTLX-9mv9vtdIQSYD4fRo7gdy_JPNuQ4_nn8x384ljoL2aTFFMCf6O7oMRObBZXaJPmT0joDmxr4RcrtGK0kXzyNmx_zK6wI1bY9ovO3jTsYJxlBBqL8lWHU3r3XXEIfT78enb00AiwrdeSlzlkSBie767DFZopCmsfM91EXrMOE3TD7TmTlr9LHyiD1ErAoLihL62preFVGKdAVTnOIJxS9OnoD81KK-zyJAFjI7Xf2suyEgJy5VNjoPPYbKSnYWXbGfM4msxfSXawKnae8sy7nF0JtGzrXtTraJw1VwIllW9Nz2j1644NGsJiexIFY66TAOUIU-UGNVGFMxulB0k_S8_HvFB4JE_EWHw34S5IeDwznwvfFQdBBpRdDG2OBW9LMztVDDuo84yBSvuHAdJGwMwLjxHWmOmXgEoOpjUDGL68n6Jb3xE01DO7DCvAanIkJEESq-IKAl1E2Ap4uo-dFCYJYxsW2pCTukMaumFYUoetZJh_7cWwAcwoZegLNCRkZ0x6CdsqL0MZznZgXlW3lysuruiUFTAtLO2hWyF2qLnv0'
    
    # Configurar headers para las peticiones
    headers = {
        'Authorization': f'Bearer {API_KEY}',
        'Content-Type': 'application/json'
    }
    
    # Probar conexiÃ³n con equipos
    try:
        response = requests.get(f"{BASE_URL}/teams", headers=headers)
        response.raise_for_status()
        teams = response.json()
        print(f"âœ… ConexiÃ³n exitosa - {len(teams)} equipos encontrados")
        return headers, BASE_URL
    except Exception as e:
        print(f"âŒ Error: {e}")
        return None

def convertir_timestamp_a_fecha(timestamp_valor):
    """Convertir timestamp a formato de fecha REAL (datetime.date) - VERSIÃ“N ROBUSTA"""
    if timestamp_valor is None:
        return None
    
    try:
        # CASO 1: Timestamp numÃ©rico (int/float)
        if isinstance(timestamp_valor, (int, float)):
            fecha = datetime.fromtimestamp(timestamp_valor)
            return fecha.date()
        
        # CASO 2: String de fecha
        if isinstance(timestamp_valor, str):
            timestamp_str = timestamp_valor.strip()
            
            # Lista de formatos de fecha a probar (del mÃ¡s especÃ­fico al mÃ¡s general)
            formatos_fecha = [
                '%Y-%m-%d',           # 2025-06-03
                '%d/%m/%Y',           # 31/05/2025
                '%m/%d/%Y',           # 05/31/2025  
                '%Y-%m-%dT%H:%M:%S',  # 2025-06-03T10:30:00
                '%Y-%m-%d %H:%M:%S',  # 2025-06-03 10:30:00
                '%d-%m-%Y',           # 31-05-2025
                '%d.%m.%Y',           # 31.05.2025
                '%Y/%m/%d',           # 2025/05/31
            ]
            
            # Intentar cada formato
            for formato in formatos_fecha:
                try:
                    # Para formatos con tiempo, extraer solo la fecha
                    if 'T' in formato or ' ' in formato:
                        fecha_datetime = datetime.strptime(timestamp_str, formato)
                        return fecha_datetime.date()
                    else:
                        # Para formatos solo de fecha
                        if len(timestamp_str) >= 8:  # MÃ­nimo para una fecha
                            # Extraer solo la parte de fecha si hay mÃ¡s contenido
                            fecha_parte = timestamp_str.split('T')[0].split(' ')[0]
                            fecha_datetime = datetime.strptime(fecha_parte, formato)
                            return fecha_datetime.date()
                except ValueError:
                    continue  # Probar siguiente formato
            
            # Si ningÃºn formato funcionÃ³, intentar parseo automÃ¡tico con dateutil si estÃ¡ disponible
            try:
                from dateutil import parser
                fecha_parseada = parser.parse(timestamp_str)
                return fecha_parseada.date()
            except ImportError:
                # dateutil no disponible, continuar con mÃ©todo manual
                pass
            except (ValueError, TypeError):
                # Error en parseo automÃ¡tico
                pass
        
        # CASO 3: Ya es un objeto datetime
        if hasattr(timestamp_valor, 'date'):
            return timestamp_valor.date()
        
        # CASO 4: Ya es un objeto date
        if hasattr(timestamp_valor, 'year') and hasattr(timestamp_valor, 'month'):
            return timestamp_valor
            
        return None
    
    except Exception as e:
        print(f"âš ï¸ Error convirtiendo fecha {timestamp_valor}: {e}")
        return None

def calcular_mts_min(total_distance, total_duration):
    """Calcular metros por minuto correctamente"""
    try:
        if total_duration is None or total_distance is None:
            return 0.0
        
        # Convertir duraciÃ³n de segundos a minutos
        if total_duration > 0:
            duration_minutes = total_duration / 60.0
            return round(total_distance / duration_minutes, 2)
        else:
            return 0.0
    except:
        return 0.0

def procesar_estadisticas_jugador(stats_row):
    """Procesar las estadÃ­sticas de un jugador EXACTAMENTE como aparece en el PDF Catapult"""
    
    # Datos bÃ¡sicos
    processed = {
        'athlete_id': stats_row.get('athlete_id', ''),
        'athlete_name': stats_row.get('athlete_name', ''),
        'activity_id': stats_row.get('activity_id', ''),
        'activity_name': stats_row.get('activity_name', ''),
        'period_id': stats_row.get('period_id', ''),
        'period_name': stats_row.get('period_name', ''),
        'date': stats_row.get('date', ''),  # â† Se procesarÃ¡ despuÃ©s
    }
    
    # â† NUEVO: Procesar fecha correctamente
    fecha_raw = stats_row.get('date', '')
    if fecha_raw:
        fecha_procesada = convertir_timestamp_a_fecha(fecha_raw)
        processed['date'] = fecha_procesada  # Ahora serÃ¡ datetime.date o None
    else:
        processed['date'] = None
    
    # DuraciÃ³n y distancia bÃ¡sicas
    total_duration = stats_row.get('total_duration', 0) or 0
    total_distance = stats_row.get('total_distance', 0) or 0
    duration_minutes = total_duration / 60.0 if total_duration > 0 else 0
    
    # NOMBRES EXACTOS DEL PDF - PÃGINA 2
    processed['Avg Dur (Sess)'] = f"{int(total_duration // 3600):02d}:{int((total_duration % 3600) // 60):02d}:{int(total_duration % 60):02d}"
    processed['Dist Sesion (m)'] = round(total_distance, 1)
    processed['Mts Min'] = calcular_mts_min(total_distance, total_duration)
    
    # Aceleraciones y desaceleraciones - NOMBRES EXACTOS DEL PDF
    acc_3_total = stats_row.get('gen2_acceleration_band8_total_effort_count', 0) or 0
    dec_3_total = stats_row.get('gen2_acceleration_band1_total_effort_count', 0) or 0
    acc_2_total = stats_row.get('gen2_acceleration_band7_total_effort_count', 0) or 0
    dec_2_total = stats_row.get('gen2_acceleration_band2_total_effort_count', 0) or 0
    
    processed['Acc +3 m/ss'] = acc_3_total
    processed['Dec +3 m/ss2'] = dec_3_total  # Nota: el PDF tiene "m/ss2" con 2 al final
    processed['A + D Effs'] = acc_3_total + dec_3_total
    
    # Player Load - NOMBRES EXACTOS DEL PDF
    player_load_total = stats_row.get('total_player_load', 0) or 0
    processed['PL Sesion'] = round(player_load_total, 1)
    processed['PL/MIN'] = round(player_load_total / duration_minutes if duration_minutes > 0 else 0, 2)
    
    # AceleraciÃ³n y desaceleraciÃ³n mÃ¡xima - NOMBRES EXACTOS DEL PDF
    processed['ACC MAX'] = round(stats_row.get('max_effort_acceleration', 0) or 0, 1)
    processed['MAX DEC'] = round(abs(stats_row.get('max_effort_deceleration', 0) or 0), 1)
    
    # Carga de aceleraciÃ³n - NOMBRE EXACTO DEL PDF
    processed['ACC LOAD'] = round(stats_row.get('total_acceleration_load', 0) or 0, 1)
    
    # Bandas de velocidad - NOMBRES EXACTOS DEL PDF
    banda_14_19 = stats_row.get('velocity_band4_total_distance', 0) or 0
    banda_19_25 = stats_row.get('velocity_band5_total_distance', 0) or 0
    banda_25_plus = stats_row.get('velocity_band6_total_distance', 0) or 0
    
    processed['Mts 14.4-19.8 k/h (m)'] = round(banda_14_19, 1)
    processed['Mts 19.8-25.2 k/h (m)'] = round(banda_19_25, 1) 
    processed['Mts +25.2km/h (m)'] = round(banda_25_plus, 1)
    processed['MAX VEL (km/h)'] = round(stats_row.get('max_vel', 0) or 0, 1)
    
    # Sprint efforts - NOMBRE EXACTO DEL PDF
    processed['Sprint +25.2km/h'] = stats_row.get('sprint_efforts', 0) or 0
    
    # MÃ‰TRICAS POR MINUTO QUE FALTAN - PÃGINA 5 DEL PDF
    if duration_minutes > 0:
        processed['Acc +3m/ss Min'] = round(acc_3_total / duration_minutes, 2)
        processed['Decc -3m/ss Min'] = round(dec_3_total / duration_minutes, 2)
        processed['Acc +2m/ss Min'] = round(acc_2_total / duration_minutes, 2)
        processed['Decc -2m/ss Min'] = round(dec_2_total / duration_minutes, 2)
        processed['Mts /min +14,4 km/h'] = round(banda_14_19 / duration_minutes, 2)
        processed['Mts / Min +19,8km/h'] = round(banda_19_25 / duration_minutes, 2)
        processed['Mts +25,2 Km/h Min'] = round(banda_25_plus / duration_minutes, 2)
    else:
        processed['Acc +3m/ss Min'] = 0.0
        processed['Decc -3m/ss Min'] = 0.0
        processed['Acc +2m/ss Min'] = 0.0
        processed['Decc -2m/ss Min'] = 0.0
        processed['Mts /min +14,4 km/h'] = 0.0
        processed['Mts / Min +19,8km/h'] = 0.0
        processed['Mts +25,2 Km/h Min'] = 0.0
    
    # MÃ©tricas adicionales que aparecen en el PDF
    processed['Meterage_Per_Minute'] = round(stats_row.get('meterage_per_minute', 0) or 0, 2)
    processed['Percentage_Max_Velocity'] = round(stats_row.get('percentage_max_velocity', 0) or 0, 1)
    processed['Explosive_Efforts'] = stats_row.get('explosive_efforts', 0) or 0
    processed['High_Speed_Efforts'] = stats_row.get('high_speed_efforts', 0) or 0
    
    return processed

# =============================================================================
# FUNCIONES DE LIMPIEZA Y NORMALIZACIÃ“N
# =============================================================================

def normalizar_nombre_jugador(nombre):
    """Normaliza la codificaciÃ³n de nombres para compatibilidad con Excel"""
    if pd.isna(nombre) or nombre == '':
        return ''
    
    nombre = str(nombre).strip()
    
    # Normalizar solo la codificaciÃ³n Unicode para compatibilidad
    nombre_normalizado = unicodedata.normalize('NFC', nombre)
    
    # Reemplazar apostrofes problemÃ¡ticos por apostrofe estÃ¡ndar
    apostrofes_problematicos = [''', ''', '`', 'Â´']
    for apostrofe_raro in apostrofes_problematicos:
        nombre_normalizado = nombre_normalizado.replace(apostrofe_raro, "'")
    
    # Limpiar solo espacios mÃºltiples y capitalizar correctamente
    nombre_final = ' '.join(nombre_normalizado.split())
    nombre_final = nombre_final.title()
    
    return nombre_final

def procesar_actividad_mejorada(actividad_str):
    """FunciÃ³n mejorada basada en patrones reales encontrados - VERSIÃ“N FINAL CORREGIDA"""
    
    if pd.isna(actividad_str) or actividad_str is None:
        return '', '', ''
    
    actividad_str = str(actividad_str).strip()
    
    if actividad_str == '' or actividad_str.lower() == 'nan':
        return '', '', actividad_str
    
    # âœ… INICIALIZAR VARIABLES - ESTO ES CLAVE PARA EVITAR EL ERROR
    equipo = ''
    jornada = ''
    actividad_limpia = actividad_str
    
    # 1. BUSCAR EQUIPO - PATRÃ“N CORREGIDO
    match_equipo = re.search(r'\b(C\d{1,2})\b', actividad_str)
    if match_equipo:
        codigo_encontrado = match_equipo.group(1)
        # Extraer nÃºmero y formatear a C##
        numero = int(re.search(r'(\d+)', codigo_encontrado).group(1))
        equipo = f"C{numero:02d}"
        
        # Remover el cÃ³digo del string
        actividad_limpia = re.sub(r'\bC\d{1,2}\b', '', actividad_limpia).strip()
    
    # 2. BUSCAR JORNADA - PATRÃ“N CORREGIDO
    match_jornada = re.search(r'FECHA\s*(\d{1,2})', actividad_limpia.upper())
    if match_jornada:
        jornada = int(match_jornada.group(1))
        # Remover "FECHA ##" del string
        actividad_limpia = re.sub(r'\s*FECHA\s*\d+', '', actividad_limpia, flags=re.IGNORECASE).strip()
    
    # 3. LIMPIAR FECHAS NUMÃ‰RICAS (DD-MM, etc.) - CORREGIDO EL REGEX
    actividad_limpia = re.sub(r'\s+\d{1,2}[-\./]\d{2}$', '', actividad_limpia).strip()
    
    # 4. LIMPIAR TIEMPOS (1T, 2T)
    actividad_limpia = re.sub(r'\s+\d+T$', '', actividad_limpia, flags=re.IGNORECASE).strip()
    
    # 5. CLASIFICAR BASADO EN PATRONES REALES
    actividad_upper = actividad_limpia.upper()
    
    if 'ENTRENAMIENTO' in actividad_upper:
        actividad_final = 'ENTRENAMIENTO'
        
    elif 'VS' in actividad_upper:
        # Extraer rival
        partes = actividad_upper.split('VS', 1)
        if len(partes) > 1:
            rival = partes[1].strip()
            # Limpiar el rival
            rival = re.sub(r'\s*\+\s*CALENTAMIENTO', '', rival)
            rival = re.sub(r'\s*\d{1,2}[-\./]\d{2}$', '', rival)  # Quitar fechas del final
            actividad_final = f'PARTIDO VS {rival.title()}'
        else:
            actividad_final = 'PARTIDO'
            
    elif 'PARTIDO' in actividad_upper:
        actividad_final = 'PARTIDO'
        
    else:
        # Si no coincide con ningÃºn patrÃ³n conocido, mantener limpio
        actividad_final = actividad_limpia.strip() or 'ACTIVIDAD'
    
    # âœ… DEVOLVER SIEMPRE LAS 3 VARIABLES INICIALIZADAS
    return equipo, jornada, actividad_final

# =============================================================================
# SISTEMA DE DETECCIÃ“N INTELIGENTE DE ESQUEMA
# =============================================================================

def obtener_mapeo_columnas():
    """Mapeo bidireccional entre nombres crudos (API) y nombres limpios (finales)"""
    mapeo_crudo_a_limpio = {
        'athlete_id': 'jugador_id',
        'athlete_name': 'nombre_jugador',
        'activity_id': 'actividad_id',
        'activity_name': 'actividad',
        'period_id': 'tarea_id',
        'period_name': 'tarea',
        'date': 'fecha'
    }
    
    # Mapeo inverso automÃ¡tico
    mapeo_limpio_a_crudo = {v: k for k, v in mapeo_crudo_a_limpio.items()}
    
    return mapeo_crudo_a_limpio, mapeo_limpio_a_crudo

def detectar_esquema_dataframe(df):
    """Detecta si un DataFrame tiene esquema CRUDO o LIMPIO"""
    if df.empty:
        return "VACIO"
    
    mapeo_crudo_a_limpio, mapeo_limpio_a_crudo = obtener_mapeo_columnas()
    
    columnas_crudas = set(mapeo_crudo_a_limpio.keys())
    columnas_limpias = set(mapeo_limpio_a_crudo.keys())
    columnas_df = set(df.columns)
    
    # Contar coincidencias
    coincidencias_crudas = len(columnas_crudas.intersection(columnas_df))
    coincidencias_limpias = len(columnas_limpias.intersection(columnas_df))
    
    if coincidencias_crudas > coincidencias_limpias:
        return "CRUDO"
    elif coincidencias_limpias > coincidencias_crudas:
        return "LIMPIO"
    else:
        # Si hay empate, revisar columnas especÃ­ficas
        if 'athlete_id' in columnas_df:
            return "CRUDO"
        elif 'jugador_id' in columnas_df:
            return "LIMPIO"
        else:
            return "DESCONOCIDO"

def normalizar_a_esquema_crudo(df, esquema_actual):
    """Convierte un DataFrame a esquema CRUDO (nombres de API) para procesamiento interno"""
    if esquema_actual == "CRUDO":
        return df.copy()
    
    if esquema_actual == "LIMPIO":
        mapeo_crudo_a_limpio, mapeo_limpio_a_crudo = obtener_mapeo_columnas()
        
        # Crear mapeo de reversiÃ³n solo para columnas que existen
        mapeo_reversion = {}
        for col_limpia in df.columns:
            if col_limpia in mapeo_limpio_a_crudo:
                mapeo_reversion[col_limpia] = mapeo_limpio_a_crudo[col_limpia]
        
        df_normalizado = df.rename(columns=mapeo_reversion)
        if mapeo_reversion:
            print(f"   ðŸ”„ Normalizando esquema LIMPIO â†’ CRUDO: {len(mapeo_reversion)} columnas")
        return df_normalizado
    
    # Si es DESCONOCIDO o VACIO, devolver como estÃ¡
    return df.copy()

def aplicar_esquema_limpio(df):
    """Aplica esquema LIMPIO (nombres finales) a un DataFrame"""
    mapeo_crudo_a_limpio, _ = obtener_mapeo_columnas()
    
    # Crear mapeo solo para columnas que existen
    mapeo_aplicar = {}
    for col_cruda in df.columns:
        if col_cruda in mapeo_crudo_a_limpio:
            mapeo_aplicar[col_cruda] = mapeo_crudo_a_limpio[col_cruda]
    
    df_limpio = df.rename(columns=mapeo_aplicar)
    
    if mapeo_aplicar:
        print(f"   âœ… Aplicando esquema LIMPIO: {len(mapeo_aplicar)} columnas renombradas")
    
    return df_limpio

def cargar_datos_existentes_inteligente():
    """Carga datos existentes y los normaliza a esquema CRUDO para procesamiento"""
    datos_existentes = {
        'jugadores': pd.DataFrame(),
        'actividades': pd.DataFrame(),
        'periodos': pd.DataFrame()
    }
    
    archivos_parquet = {
        'jugadores': 'data/catapult_jugadores.parquet',
        'actividades': 'data/catapult_actividades.parquet',
        'periodos': 'data/catapult_periodos.parquet'
    }
    
    for tipo, archivo in archivos_parquet.items():
        if os.path.exists(archivo):
            try:
                df = pd.read_parquet(archivo)
                esquema = detectar_esquema_dataframe(df)
                
                print(f"   ðŸ“„ {archivo}: {len(df)} registros, esquema {esquema}")
                
                # Normalizar a esquema crudo para procesamiento interno
                df_normalizado = normalizar_a_esquema_crudo(df, esquema)
                datos_existentes[tipo] = df_normalizado
                
            except Exception as e:
                print(f"   âŒ Error cargando {archivo}: {e}")
        else:
            print(f"   ðŸ“„ {archivo}: No existe, se crearÃ¡ nuevo")
    
    return datos_existentes

def obtener_ids_existentes_flexible(datos_existentes):
    """Extraer IDs de manera flexible (funciona con esquemas crudo y limpio)"""
    ids_existentes = {
        'jugadores': set(),
        'actividades': set(),
        'periodos': set()
    }
    
    # Posibles nombres de columnas ID (crudo y limpio)
    posibles_columnas_id = {
        'jugadores': ['athlete_id', 'jugador_id'],
        'actividades': ['activity_id', 'actividad_id', 'id'],
        'periodos': ['period_id', 'tarea_id', 'id']
    }
    
    for tipo, df in datos_existentes.items():
        if not df.empty:
            for posible_col in posibles_columnas_id[tipo]:
                if posible_col in df.columns:
                    ids_existentes[tipo] = set(df[posible_col].dropna().unique())
                    print(f"   ðŸ”‘ {tipo}: {len(ids_existentes[tipo])} IDs desde columna '{posible_col}'")
                    break
    
    return ids_existentes

def limpiar_dataframe_final(df, tipo_datos):
    """Limpieza integral final que se aplica DESPUÃ‰S de combinar datos existentes + nuevos"""
    if df.empty:
        return df
    
    print(f"ðŸ§¹ Aplicando limpieza final a {tipo_datos}...")
    
    # Hacer copia para trabajar
    df_limpio = df.copy()
    
    # PASO 1: Aplicar esquema limpio (renombrar columnas) - SOLO SI NO ESTÃ YA APLICADO
    columnas_limpias = ['jugador_id', 'nombre_jugador', 'actividad_id', 'actividad', 'tarea_id', 'tarea', 'fecha']
    tiene_esquema_limpio = any(col in df_limpio.columns for col in columnas_limpias)
    
    if not tiene_esquema_limpio:
        df_limpio = aplicar_esquema_limpio(df_limpio)
    
    # PASO 2: Normalizar nombres de jugadores (siempre seguro)
    if 'nombre_jugador' in df_limpio.columns:
        df_limpio['nombre_jugador'] = df_limpio['nombre_jugador'].apply(normalizar_nombre_jugador)
    
    # PASO 3: Procesar columna actividad SOLO SI NO EXISTEN ya las columnas equipo y jornada
    ya_tiene_equipo_jornada = 'equipo' in df_limpio.columns and 'jornada' in df_limpio.columns
    
    if 'actividad' in df_limpio.columns and not ya_tiene_equipo_jornada:
        print(f"   ðŸ”„ Extrayendo equipo y jornada (primera vez)...")
        df_limpio['equipo'] = ''
        df_limpio['jornada'] = ''
        
        for idx in range(len(df_limpio)):
            try:
                actividad = df_limpio['actividad'].iloc[idx]
                equipo, jornada, actividad_nueva = procesar_actividad_individual(actividad)
                
                df_limpio.loc[df_limpio.index[idx], 'equipo'] = equipo
                df_limpio.loc[df_limpio.index[idx], 'jornada'] = jornada
                df_limpio.loc[df_limpio.index[idx], 'actividad'] = actividad_nueva
                
            except Exception as e:
                df_limpio.loc[df_limpio.index[idx], 'equipo'] = ''
                df_limpio.loc[df_limpio.index[idx], 'jornada'] = ''
    elif ya_tiene_equipo_jornada:
        print(f"   âœ… Equipo y jornada ya existen, manteniendo valores existentes")
    
    # PASO 4: ELIMINAR COLUMNAS COMPLETAMENTE VACÃAS (CORREGIDO - SIN ELIMINAR COLUMNAS CON CEROS)
    columnas_iniciales = len(df_limpio.columns)
    columnas_vacias = []
    
    for columna in df_limpio.columns:
        # Verificar si la columna estÃ¡ completamente vacÃ­a
        serie = df_limpio[columna]
        
        # CONDICIONES CORREGIDAS - NO eliminar columnas con solo ceros
        if (serie.isna().all() or                                    # Todos NaN
            serie.isnull().all() or                                  # Todos None
            (serie.astype(str).str.strip() == '').all()):           # Todos strings vacÃ­os
            # REMOVIDO: (serie == 0).all() - LAS MÃ‰TRICAS PUEDEN SER LEGÃTIMAMENTE 0
            
            columnas_vacias.append(columna)
    
    # Eliminar columnas vacÃ­as (excepto las importantes para la estructura)
    columnas_importantes = [
        'jugador_id', 'nombre_jugador', 'actividad_id', 'actividad', 
        'tarea_id', 'tarea', 'fecha', 'equipo', 'jornada'
    ]
    
    columnas_a_eliminar = [col for col in columnas_vacias if col not in columnas_importantes]
    
    if columnas_a_eliminar:
        print(f"   ðŸ—‘ï¸ Eliminando {len(columnas_a_eliminar)} columnas realmente vacÃ­as:")
        for col in columnas_a_eliminar[:5]:  # Mostrar solo las primeras 5
            print(f"      â€¢ {col}")
        if len(columnas_a_eliminar) > 5:
            print(f"      â€¢ ... y {len(columnas_a_eliminar) - 5} mÃ¡s")
        
        df_limpio = df_limpio.drop(columns=columnas_a_eliminar)
    
    # PASO 5: Eliminar duplicados finales - CON VALIDACIÃ“N MEJORADA
    filas_inicial = len(df_limpio)
    df_limpio = df_limpio.drop_duplicates()
    
    # Eliminar duplicados por ID especÃ­ficos - SOLO SI LAS COLUMNAS EXISTEN
    columnas_id_por_tipo = {
        'JUGADORES': ['jugador_id', 'tarea_id', 'actividad_id'],
        'ACTIVIDADES': ['actividad_id'],
        'PERÃODOS': ['tarea_id', 'actividad_id']
    }
    
    columnas_id = columnas_id_por_tipo.get(tipo_datos.upper(), [])
    columnas_existentes = [col for col in columnas_id if col in df_limpio.columns]
    
    # VALIDACIÃ“N: Solo eliminar duplicados si hay al menos UNA columna ID vÃ¡lida
    if columnas_existentes and len(df_limpio) > 0:
        filas_antes_duplicados = len(df_limpio)
        df_limpio = df_limpio.drop_duplicates(subset=columnas_existentes, keep='last')
        duplicados_eliminados = filas_antes_duplicados - len(df_limpio)
        if duplicados_eliminados > 0:
            print(f"   ðŸ”„ Eliminados {duplicados_eliminados} duplicados por ID")
    
    filas_final = len(df_limpio)
    
    # PASO 6: Reordenar columnas
    columnas_principales = [
        'jugador_id', 'nombre_jugador', 'equipo', 'jornada', 
        'actividad_id', 'actividad', 'tarea_id', 'tarea', 'fecha'
    ]
    
    columnas_existentes = [col for col in columnas_principales if col in df_limpio.columns]
    otras_columnas = [col for col in df_limpio.columns if col not in columnas_principales]
    
    nuevo_orden = columnas_existentes + otras_columnas
    df_limpio = df_limpio[nuevo_orden]
    
    # PASO 7: Eliminar columnas especÃ­ficas para actividades y perÃ­odos - SOLO SI ES NECESARIO
    if tipo_datos.upper() in ["ACTIVIDADES", "PERÃODOS"]:
        columnas_a_eliminar_final = ['jugador_id', 'nombre_jugador']
        columnas_presentes = [col for col in columnas_a_eliminar_final if col in df_limpio.columns]
        if columnas_presentes:
            df_limpio = df_limpio.drop(columns=columnas_presentes)
            print(f"   ðŸ—‚ï¸ Eliminadas columnas de jugador para {tipo_datos}: {columnas_presentes}")
    
    # RESUMEN FINAL
    columnas_finales = len(df_limpio.columns)
    columnas_eliminadas_total = columnas_iniciales - columnas_finales
    
    print(f"   âœ… {tipo_datos}: {filas_inicial} â†’ {filas_final} registros (-{filas_inicial-filas_final} duplicados)")
    print(f"   ðŸ“Š Columnas: {columnas_iniciales} â†’ {columnas_finales} (-{columnas_eliminadas_total} vacÃ­as)")
    
    # VALIDACIÃ“N FINAL: Asegurarse de que no se hayan eliminado TODAS las filas
    if filas_inicial > 0 and filas_final == 0:
        print(f"   âŒ ERROR CRÃTICO: Se eliminaron TODAS las filas en {tipo_datos}")
        print(f"   ðŸ”„ Devolviendo DataFrame original para evitar pÃ©rdida de datos")
        return df  # Devolver DataFrame original si algo saliÃ³ mal
    
    return df_limpio

def limpiar_datos_nuevos_solamente(df_nuevos, tipo_datos):
    """Limpieza que se aplica SOLO a datos nuevos antes de combinar con existentes"""
    if df_nuevos.empty:
        return df_nuevos
    
    print(f"ðŸ§¹ Limpiando datos NUEVOS de {tipo_datos}...")
    
    # Hacer copia para trabajar
    df_limpio = df_nuevos.copy()
    
    # PASO 1: Aplicar esquema limpio (renombrar columnas)
    df_limpio = aplicar_esquema_limpio(df_limpio)
    
    # PASO 2: Normalizar nombres de jugadores
    if 'nombre_jugador' in df_limpio.columns:
        df_limpio['nombre_jugador'] = df_limpio['nombre_jugador'].apply(normalizar_nombre_jugador)
    
    # PASO 3: Procesar columna actividad (extraer equipo, jornada) - SOLO PARA NUEVOS
    if 'actividad' in df_limpio.columns:
        df_limpio['equipo'] = ''
        df_limpio['jornada'] = ''
        
        for idx in range(len(df_limpio)):
            try:
                actividad = df_limpio['actividad'].iloc[idx]
                equipo, jornada, actividad_nueva = procesar_actividad_individual(actividad)
                
                df_limpio.loc[df_limpio.index[idx], 'equipo'] = equipo
                df_limpio.loc[df_limpio.index[idx], 'jornada'] = jornada
                df_limpio.loc[df_limpio.index[idx], 'actividad'] = actividad_nueva
                
            except Exception as e:
                df_limpio.loc[df_limpio.index[idx], 'equipo'] = ''
                df_limpio.loc[df_limpio.index[idx], 'jornada'] = ''
    
    print(f"   âœ… Datos nuevos de {tipo_datos} limpiados: {len(df_limpio)} registros")
    return df_limpio

def guardar_dataframes_limpios_finales(df_activities, df_periods, df_players):
    """FunciÃ³n CORREGIDA - Aplica limpieza final a DataFrames en esquema CRUDO"""
    
    print("ðŸ§¹ Aplicando limpieza final (desde esquema CRUDO a LIMPIO)...")
    
    def limpieza_final_crudo_a_limpio(df, tipo_datos):
        """Limpieza que convierte de esquema CRUDO a LIMPIO - CORREGIDA"""
        if df.empty:
            return df
        
        print(f"   ðŸ”„ Procesando {tipo_datos}...")
        df_limpio = df.copy()
        filas_inicial = len(df_limpio)
        
        # PASO 1: Aplicar esquema LIMPIO (renombrar columnas de API a nombres finales)
        df_limpio = aplicar_esquema_limpio(df_limpio)
        
        # PASO 2: Normalizar nombres de jugadores (solo si existe la columna)
        if 'nombre_jugador' in df_limpio.columns:
            df_limpio['nombre_jugador'] = df_limpio['nombre_jugador'].apply(normalizar_nombre_jugador)
        
        # PASO 3: Procesar actividad (CORREGIDO - procesar registros con equipo/jornada vacÃ­os)
        if 'actividad' in df_limpio.columns:
            # Asegurar que existan las columnas
            if 'equipo' not in df_limpio.columns:
                df_limpio['equipo'] = ''
            if 'jornada' not in df_limpio.columns:
                df_limpio['jornada'] = ''
            
            # CLAVE: Procesar registros donde equipo O jornada estÃ¡n vacÃ­os (datos nuevos)
            registros_a_procesar = (
                (df_limpio['equipo'].isna()) | 
                (df_limpio['equipo'] == '') | 
                (df_limpio['jornada'].isna()) | 
                (df_limpio['jornada'] == '')
            )
            
            registros_nuevos = registros_a_procesar.sum()
            if registros_nuevos > 0:
                print(f"      ðŸ”„ Procesando {registros_nuevos} registros nuevos (equipo/jornada vacÃ­os)...")
                
                for idx in df_limpio[registros_a_procesar].index:
                    try:
                        actividad = df_limpio.loc[idx, 'actividad']
                        equipo, jornada, actividad_nueva = procesar_actividad_individual(actividad)
                        
                        df_limpio.loc[idx, 'equipo'] = equipo
                        df_limpio.loc[idx, 'jornada'] = jornada
                        df_limpio.loc[idx, 'actividad'] = actividad_nueva
                        
                    except Exception as e:
                        df_limpio.loc[idx, 'equipo'] = ''
                        df_limpio.loc[idx, 'jornada'] = ''
            else:
                print(f"      âœ… Todos los registros ya tienen equipo/jornada procesados")
        
        # PASO 4: Eliminar duplicados exactos
        df_limpio = df_limpio.drop_duplicates()
        duplicados_exactos = filas_inicial - len(df_limpio)
        
        # PASO 5: Reordenar columnas
        columnas_principales = [
            'jugador_id', 'nombre_jugador', 'equipo', 'jornada', 
            'actividad_id', 'actividad', 'tarea_id', 'tarea', 'fecha'
        ]
        
        columnas_existentes_orden = [col for col in columnas_principales if col in df_limpio.columns]
        otras_columnas = [col for col in df_limpio.columns if col not in columnas_principales]
        
        if columnas_existentes_orden:
            nuevo_orden = columnas_existentes_orden + otras_columnas
            df_limpio = df_limpio[nuevo_orden]
        
        # PASO 6: Eliminar columnas de jugador para actividades y perÃ­odos
        if tipo_datos.upper() in ["ACTIVIDADES", "PERÃODOS"]:
            columnas_a_eliminar = ['jugador_id', 'nombre_jugador']
            columnas_presentes = [col for col in columnas_a_eliminar if col in df_limpio.columns]
            if columnas_presentes:
                df_limpio = df_limpio.drop(columns=columnas_presentes)
        
        filas_final = len(df_limpio)
        
        if duplicados_exactos > 0:
            print(f"      ðŸ”„ {tipo_datos}: {filas_inicial} â†’ {filas_final} registros (-{duplicados_exactos} duplicados)")
        else:
            print(f"      âœ… {tipo_datos}: {filas_inicial} registros procesados")
        
        return df_limpio
    
    # Aplicar limpieza final a todos los DataFrames
    df_activities_limpio = limpieza_final_crudo_a_limpio(df_activities, 'ACTIVIDADES')
    df_periods_limpio = limpieza_final_crudo_a_limpio(df_periods, 'PERÃODOS')  
    df_players_limpio = limpieza_final_crudo_a_limpio(df_players, 'JUGADORES')
    
    # Crear carpeta data si no existe
    os.makedirs('data', exist_ok=True)

    # Guardar con nombres finales limpios
    archivos = [
        ("data/catapult_actividades.parquet", df_activities_limpio, "ACTIVIDADES"),
        ("data/catapult_periodos.parquet", df_periods_limpio, "PERÃODOS"),
        ("data/catapult_jugadores.parquet", df_players_limpio, "JUGADORES")
    ]
    
    print("\nðŸ’¾ Guardando DataFrames limpios...")
    
    for filename, df, label in archivos:
        try:
            if df is not None and not df.empty:
                df.to_parquet(filename, index=False)
                size_mb = os.path.getsize(filename) / 1024 / 1024
                print(f"âœ… {label}: {filename} guardado ({len(df):,} registros, {size_mb:.1f} MB)")
            else:
                print(f"âš ï¸ {label}: DataFrame vacÃ­o, no se guarda")
                
        except Exception as e:
            print(f"âŒ Error guardando {label}: {e}")

def convertir_tipos_finales(df):
    """Convertir tipos de datos finales para asegurar consistencia - VERSIÃ“N ROBUSTA"""
    if df.empty:
        return df
    
    df_convertido = df.copy()
    
    # 1. CONVERTIR FECHAS (usando la funciÃ³n robusta existente)
    if 'fecha' in df_convertido.columns:
        print(f"    ðŸ”„ Convirtiendo fechas a datetime.date...")
        
        # Usar la funciÃ³n robusta que ya maneja mÃºltiples formatos
        df_convertido['fecha'] = df_convertido['fecha'].apply(convertir_timestamp_a_fecha)
        
        # Contar Ã©xitos
        fechas_validas = df_convertido['fecha'].notna().sum()
        fechas_totales = len(df_convertido)
        print(f"    ðŸ“… {fechas_validas}/{fechas_totales} fechas convertidas exitosamente")
    
    # 2. CONVERTIR JORNADAS
    if 'jornada' in df_convertido.columns:
        print(f"    ðŸ”¢ Convirtiendo jornadas a enteros...")
        
        def convertir_jornada_segura(jornada_val):
            if pd.isna(jornada_val) or jornada_val is None:
                return None
            
            try:
                # Convertir a string y limpiar
                jornada_str = str(jornada_val).strip()
                
                # Si estÃ¡ vacÃ­o despuÃ©s del strip
                if jornada_str == '' or jornada_str.lower() == 'nan':
                    return None
                
                # Remover ceros a la izquierda y convertir a int
                jornada_int = int(jornada_str.lstrip('0') or '0')
                return jornada_int
                
            except (ValueError, TypeError, AttributeError):
                return None
        
        df_convertido['jornada'] = df_convertido['jornada'].apply(convertir_jornada_segura)
        
        # Contar Ã©xitos
        jornadas_validas = df_convertido['jornada'].notna().sum()
        jornadas_totales = len(df_convertido)
        print(f"    ðŸ”¢ {jornadas_validas}/{jornadas_totales} jornadas convertidas exitosamente")
    
    # 3. LIMPIAR EQUIPOS (asegurar formato C##)
    if 'equipo' in df_convertido.columns:
        print(f"    ðŸˆ Normalizando equipos a formato C##...")
        
        def limpiar_equipo_final(equipo_val):
            if pd.isna(equipo_val) or equipo_val is None:
                return ''
            
            equipo_str = str(equipo_val).strip()
            
            # Si estÃ¡ vacÃ­o
            if equipo_str == '' or equipo_str.lower() == 'nan':
                return ''
            
            # Si ya estÃ¡ en formato C## correcto, mantenerlo
            if re.match(r'^C\d{2}$', equipo_str):
                return equipo_str
            
            # Si estÃ¡ en formato C# (un dÃ­gito), convertir a C##
            match = re.match(r'^C(\d{1,2})$', equipo_str)
            if match:
                numero = int(match.group(1))
                return f"C{numero:02d}"
            
            # Si no coincide con patrÃ³n, devolver como estÃ¡
            return equipo_str
        
        df_convertido['equipo'] = df_convertido['equipo'].apply(limpiar_equipo_final)
        
        # Contar equipos vÃ¡lidos
        equipos_validos = (df_convertido['equipo'] != '').sum()
        equipos_totales = len(df_convertido)
        print(f"    ðŸˆ {equipos_validos}/{equipos_totales} equipos normalizados exitosamente")
    
    return df_convertido
```

```python
import pandas as pd

# DataFrame ficticio SIN la columna 'Entradas'
data = {
    'Atleta': ['Juan PÃ©rez', 'Lucas GÃ³mez', 'MartÃ­n DÃ­az', 'Pedro LÃ³pez'],
    'Promedio Back in Game (mm:ss)': ['00:45', '01:10', '00:55', '01:25'],
    'Tiempo Activo (mm:ss)': ['30:00', '28:15', '32:10', '27:45']
}
df_resumen = pd.DataFrame(data)

# FunciÃ³n para convertir mm:ss a segundos
def mmss_a_seg(mmss):
    minutos, segundos = map(int, mmss.split(':'))
    return minutos * 60 + segundos

# Calcular promedio de 'Promedio Back in Game (mm:ss)'
promedios = df_resumen['Promedio Back in Game (mm:ss)'].apply(mmss_a_seg)
promedio_general = promedios.mean()

def color_promedio(val):
    valor = mmss_a_seg(val)
    if valor < promedio_general:
        return 'background-color: #ffcccc; color: #b71c1c; font-weight: bold;'  # rojo
    elif valor == promedio_general:
        return 'background-color: #fff7cc; color: #ff6f00; font-weight: bold;'  # amarillo
    else:
        return 'background-color: #ccffcc; color: #1b5e20; font-weight: bold;'  # verde

def nombre_grande(val):
    return 'font-size: 18px; font-weight: bold; color: #1565c0; background-color: #e3f2fd;'

# Aplicar barra horizontal a 'Tiempo Activo (mm:ss)' segÃºn su valor
def barra_tiempo_activo(val):
    # Normalizar el valor entre 0 y 1 para la barra
    valores = df_resumen['Tiempo Activo (mm:ss)'].apply(mmss_a_seg)
    minimo = valores.min()
    maximo = valores.max()
    actual = mmss_a_seg(val)
    ancho = (actual - minimo) / (maximo - minimo) if maximo > minimo else 0
    color = "#07b2e6"
    return f"background: linear-gradient(90deg, {color} {ancho*100:.0f}%, transparent {ancho*100:.0f}%);"

# Aplicar estilos
styled_df = (
    df_resumen.style
    .applymap(nombre_grande, subset=['Atleta'])
    .applymap(color_promedio, subset=['Promedio Back in Game (mm:ss)'])
    .applymap(barra_tiempo_activo, subset=['Tiempo Activo (mm:ss)'])
    .set_properties(**{'border': '2px solid #1976d2'}, subset=pd.IndexSlice[:, :])
    .set_caption("ðŸ‰ <b>Resumen Back in Game - Demo Visual</b>")
)

styled_df
```
